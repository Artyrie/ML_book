```python
""" 샘플 제작 """
import numpy as np
import pandas as pd

np.random.seed(123)
variables = ['X', 'Y', 'Z']
labels = ['ID_0', 'ID_1', 'ID_2', 'ID_3', 'ID_4']
X = np.random.random_sample([5, 3]) * 10
df = pd.DataFrame(X, columns = variables, index = labels)
df
```

```python
""" 거리 행렬 계산 """
from scipy.spatial.distance import pdist, squareform
# pdist는 축약된 거리 행렬을 반환
# squareform에 넣어 샘플 간 거리 대칭 행렬 제작

row_dist = pd.DataFrame(squareform(pdist(df, metric = 'euclidean')),
                       columns = labels, index = labels)
row_dist
```

```python
""" 연결 행렬 함수 살펴보기 """
from scipy.cluster.hierarchy import linkage
help(linkage)

print('상삼각행렬이나 값만을 전달하여 인자로 사용해야함. squareform X. 405p')
```

```python
""" 연결 함수 계산 """

row_clusters = linkage(df.values, method = 'complete', metric = 'euclidean')
# row_clusters = linkage(pdist(df, metric = 'euclidean'), method = 'complete')
pd.DataFrame(row_clusters,
            columns = ['row label 1', 'row label 2', 'distance', 'no. of items in clust.'],
            index = ['cluster %d' %(i+1) for i in range(row_clusters.shape[0])])
```

```python
""" 덴드로그램 그리기 """
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt

row_dendr = dendrogram(row_clusters, labels = labels)
plt.tight_layout()
plt.ylabel('Euclidean distance')
plt.show()
```

```python
""" 단색으로 그리기 """
from scipy.cluster.hierarchy import set_link_color_palette

set_link_color_palette(['black'])
row_dendr2 = dendrogram(row_clusters, labels = labels,
                       color_threshold = np.inf)
plt.tight_layout()
plt.ylabel('Euclidean distance')
plt.show()
```

```python
""" 히트맵에 덴드로그램 연결 """
#  자세한 내용은 408p

# 새 figure 객체를 만들고 x, y축 위치와 너비 높이 지정 후 덴드로그램 회전
fig = plt.figure(figsize = (8, 8), facecolor = 'white')
axd = fig.add_axes([0.09, 0.1, 0.2, 0.6])
row_dendr = dendrogram(row_clusters, orientation = 'left')

# 클러스터 레이블에 따라 DataFrame 데이터 재정렬
df_rowclust = df.iloc[row_dendr['leaves'][::-1]]

# 히트맵 제작 및 덴드로그램 위치잡기
axm = fig.add_axes([0.23, 0.1, 0.6, 0.6])
cax = axm.matshow(df_rowclust,
                 interpolation = 'nearest', cmap = 'hot_r')

# 축 눈금, 그래프 테두리 제거. 컬러막대 추가
axd.set_xticks([])
axd.set_yticks([])

for i in axd.spines.values() : 
    i.set_visible(False)

fig.colorbar(cax)
axm.set_xticklabels([''] + list(df_rowclust.columns))
axm.set_yticklabels([''] + list(df_rowclust.index))
plt.show()
```

```python
""" 사이킷 런의 병합 계층 군집 """
from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(n_clusters = 3,
                           affinity = 'euclidean',
                           linkage = 'complete')
labels = ac.fit_predict(X)
print('클러스터 레이블 : %s' % labels)
```

```python
""" 클러스터 개수를 줄여 샘플 3의 관계 파악 """

ac = AgglomerativeClustering(n_clusters = 2,
                           affinity = 'euclidean',
                           linkage = 'complete')
labels = ac.fit_predict(X)
print('클러스터 레이블 : %s' % labels)
```
