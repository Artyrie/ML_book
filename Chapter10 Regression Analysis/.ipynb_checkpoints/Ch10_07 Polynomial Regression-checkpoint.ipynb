```python
""" 2차 다항식 항 추가하기 """
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([258.0, 270.0, 294.0, 320.0, 342.0,
             368.0, 396.0, 446.0, 480.0, 586.0])[:, np.newaxis]
y = np.array([236.4, 234.4, 252.8, 298.6, 314.2,
             342.2, 360.8, 368.0, 391.2, 390.8])

lr = LinearRegression()
pr = LinearRegression()
quadratic = PolynomialFeatures(degree = 2)
X_quad = quadratic.fit_transform(X)
```

```python
""" 비교를 위한 선형 회귀 모델 훈련 """

lr.fit(X, y)
X_fit = np.arange(250, 600, 10)[:, np.newaxis]
y_lin_fit = lr.predict(X_fit)
```

```python
""" 다변량 회귀 모델 훈련 """

pr.fit(X_quad, y)
y_quad_fit = pr.predict(quadratic.fit_transform(X_fit))
```

```python
""" 결과 그래프 그리기 """
import matplotlib.pyplot as plt

plt.scatter(X, y, label = 'training points')
plt.plot(X_fit, y_lin_fit, label = 'linear fit', linestyle = '--')
plt.plot(X_fit, y_quad_fit, label = 'quadratic fit')
plt.legend(loc = 'upper left')
plt.show()
```

```python
""" MSE, R2 비교 """
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

y_lin_pred = lr.predict(X)
y_quad_pred = pr.predict(X_quad)
print('훈련 MSE 비교 - 선형 모델 : %.3f, 다항 모델 : %.3f' % (
    mean_squared_error(y, y_lin_pred), mean_squared_error(y, y_quad_pred)))
print('훈련 R2 비교 - 선형 모델 : %.3f, 다항 모델 : %.3f' % (
    r2_score(y, y_lin_pred), r2_score(y, y_quad_pred)))
```

```python
""" 하우징 데이터 정리 """
import pandas as pd

df = pd.read_csv('../dataset/housing.data', sep='\s+')
df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE',
             'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

X = df[['LSTAT']].values
y = df['MEDV'].values
```

```python
""" 주택 데이터셋을 사용한 비선형 관계 모델링 """

regr = LinearRegression()

# 2차, 3차 다항식 특성 만들기
quadratic = PolynomialFeatures(degree = 2)
cubic = PolynomialFeatures(degree = 3)

X_quad = quadratic.fit_transform(X)
X_cubic = cubic.fit_transform(X)

# 학습된 모델을 그리기 위한 특성 범위 만들기
X_fit = np.arange(X.min(), X.max(), 1)[:, np.newaxis]

regr = regr.fit(X, y)
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y, regr.predict(X))

regr = regr.fit(X_quad, y)
y_quad_fit = regr.predict(quadratic.fit_transform(X_fit))
quadratic_r2 = r2_score(y, regr.predict(X_quad))

regr = regr.fit(X_cubic, y)
y_cubic_fit = regr.predict(cubic.fit_transform(X_fit))
cubic_r2 = r2_score(y, regr.predict(X_cubic))
```

```python
""" 결과 그래프 그리기 """

plt.scatter(X, y, label = 'training points', color = 'lightgray')

plt.plot(X_fit, y_lin_fit,
        label = 'linear, $R2 = %.2f$' % linear_r2,
        color = 'blue', linewidth = 2, linestyle = ':')
plt.plot(X_fit, y_quad_fit,
        label = 'quadratic, $R2 = %.2f$' % quadratic_r2,
        color = 'red', linewidth = 2, linestyle = '-')
plt.plot(X_fit, y_cubic_fit,
        label = 'cubic, $R2 = %.2f$' % cubic_r2,
        color = 'green', linewidth = 2, linestyle = '--')
plt.xlabel('% lower status of the population [LSTAT]')
plt.ylabel('Price in $1000s [MEDV]')
plt.legend(loc = 'upper right')
plt.show()
```

```python
""" 위 데이터를 선형 모델로 만들기 """
# 가정은 376p 참고

# 특성 변환
X_log = np.log(X)
y_sqrt = np.sqrt(y)

# 학습된 모델을 그리기 위해 특성 범위를 만들기
X_fit = np.arange(X_log.min() - 1, X_log.max() + 1, 1)[:, np.newaxis]
regr = regr.fit(X_log, y_sqrt)
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y_sqrt, regr.predict(X_log))

# 결과 그래프 그리기
plt.scatter(X_log, y_sqrt, label = 'training points', color = 'lightgray')
plt.plot(X_fit, y_lin_fit,
        label = 'linear, $R2 = %.2f$' % linear_r2,
        color = 'blue', linewidth = 2)
plt.xlabel('log(% lower status of the population [LSTAT])')
plt.ylabel('$\sqrt{Price \; in \; \$1000s \; [MEDV]}$')
plt.legend(loc = 'lower left')
plt.show()

print('cubic 다항 특성보다 좋은 성능을 보인다.')
```
