```python
## 데이터셋 제작
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

def make_random_data() : 
    x = np.random.uniform(low = -2, high = 2, size = 200)
    y = []
    for t in x : 
        r = np.random.normal(loc = 0.0, scale = (0.5 + t * t / 3), size = None)
        y.append(r)
    return x, 1.726 * x - 0.84 + np.array(y)

x, y = make_random_data()

plt.plot(x, y, 'o')
plt.show()
```

```python
## 데이터 분할

x_train, y_train = x[:150], y[:150]
x_test, y_test = x[150:], y[150:]
```

```python
## Model 제작
import tensorflow as tf

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1, input_dim = 1))

# 유닛이 1개이므로 파라미터 개수는 가중치와 절편 두 개뿐이다.
model.summary()
```

```python
## 컴파일

model.compile(optimizer = 'sgd', loss = 'mse')
history = model.fit(x_train, y_train, epochs = 500, validation_split = 0.3)
```

```python
## 손실 그래프

epochs = np.arange(1, 500 + 1)
plt.plot(epochs, history.history['loss'], label = 'Training loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
## 함수형 API
from tensorflow.keras import Model, Input

input = tf.keras.Input(shape = (1,))
output = tf.keras.layers.Dense(1)(input)

#dense = tf.keras.layers.Dense(1)
#output = dense(input)

model = tf.keras.Model(input, output)
# Sequential 모델은 자동으로 InputLayer를 추가시켜준다.

model.summary()
```

```python
model.compile(optimizer = 'sgd', loss = 'mse')
history = model.fit(x_train, y_train, epochs = 500, validation_split = 0.3)
```

```python
epochs = np.arange(1, 500 + 1)
plt.plot(epochs, history.history['loss'], label = 'Training loss')
plt.plot(epochs, history.history['val_loss'], label = 'Validationloss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
## 모델의 가중치 저장과 복원

model.save_weights('simple_weights.h5')
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1, input_dim = 1))
model.compile(optimizer = 'sgd', loss = 'mse')
model.load_weights('simple_weights.h5')
```

```python
## 테스트 세트 평가법

model.evaluate(x_test, y_test)
```

```python
## 모델 저장

model.save('simple_model.h5')
```

```python
model = tf.keras.models.load_model('simple_model.h5')
model.evaluate(x_test, y_test)
```

```python
## Callback 사용해보기

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1, input_dim = 1))
model.compile(optimizer = 'sgd', loss = 'mse')

callback_list = [tf.keras.callbacks.ModelCheckpoint(filepath = 'my_model.h5',
                                                   monitor = 'val_loss', save_best_only = True),
                tf.keras.callbacks.EarlyStopping(patience = 5)]

history = model.fit(x_train, y_train, epochs = 500,
                    validation_split = 0.2, callbacks = callback_list)
```

```python
epochs = np.arange(1, len(history.history['loss']) + 1)
plt.plot(epochs, history.history['loss'], label = 'Training loss')
plt.plot(epochs, history.history['val_loss'], label = 'Validationloss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig('test_graph.png', dpi = 300)
plt.show()
```

```python
## 최상의 가중치 유지하기

model = tf.keras.models.load_model('simple_model.h5')
model.load_weights('my_model.h5')
model.evaluate(x_test, y_test)
```

```python
## 학습한 그래프 확인

x_arr = np.arange(-2, 2, 0.1)
y_arr = model.predict(x_arr)

plt.figure()
plt.plot(x_train, y_train, 'bo')
plt.plot(x_test, y_test, 'bo', alpha = 0.3)
plt.plot(x_arr, y_arr, '-r', lw = 3)
plt.show()
```
